```python
#1Q. How many peoples have heart disease and how many people doesn't have disease?
#Getting the values.
da.target.value_counts()
```

```python
#plotting bar chart.
da.target.value_counts().plot(kind= 'bar', color=["orchid", "salmon"])
plt.title("Heart Disease values")
plt.xlabel("1 = Heart Disease, 0 = No heart Disease")
plt.ylabel("Amount")
```

```python
#plotting a pie chart
da.target.value_counts().plot(kind='pie', figsize = (8, 6))
plt.legend(["Disease", "No Disease"])
```

```python
#'0' represent female

#'1' represent male

#SEX column part

#'0' represent 'No disease'

#'1' represent 'disease'

#Target column part

#now lets check how may of males and female are in the dataset.
da.sex.value_counts()
```

```python
#plotting a pic chart
da.sex.value_counts().plot(kind='pie', figsize=(8, 6))
plt.title("Male and Females Ratio")
plt.legend(['Male', 'Females'])
```

```python
#let's find the answers of our 2nd questions
#2Q. People of which sex has most heart disease?
pd.crosstab(da.target, da.sex)
```

```python
sns.countplot(x ='target', data = da, hue='sex')
plt.title("Heart Disease Frequency for sex")
plt.xlabel("0 = No heart Disease, 1= Heart Disease")
```

```python
#3Q. People of which sex has which type of chest pain most?

# counting values for different cheast pain most?

#counting values for different chest pain
da.cp.value_counts() 
```

```python
#plotting a bar chart
da.cp.value_counts().plot(kind='bar', color=['salmon', 'lightskyblue', 'springgreen', 'khaki'])
plt.title("chest pain types vs counts")
```

```python
pd.crosstab(da.sex, da.cp)
```

```python
pd.crosstab(da.sex, da.cp).plot(kind= 'bar', color = ['coral', 'lightskyblue', 'plum','khaki'])
plt.title('Type of chest pain for sex')
plt.xlabel('0 = Female,  1=Male')
```
