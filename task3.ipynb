```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data=pd.read_csv("C:\\Users\\kittu\\Downloads\\householdtask3.csv")
```

```python
display(data.head(10))
```

```python
#scatter plot with  year against own
plt.scatter(data['year'], data['own'])

#Adding title to the plot
plt.title("Scatter plot")

#setting the x and y label
plt.xlabel("year")
plt.ylabel('own')

#showing the result
plt.show()
```

```python
#line chart with year against owm
plt.plot(data['year'])
plt.plot(data['own'])
#Adding title to the plot
plt.title("Line chat")

#setting the x and y label
plt.xlabel("year")
plt.ylabel('own')

#showing the result
plt.show()
```

```python
#bar chart or bar plot
plt.bar(data['year'], data['own'])

#Adding title to the plot
plt.title("Bar chart")

#setting the x and y label
plt.xlabel("year")
plt.ylabel('own')

#showing the result
plt.show()
```

```python
#histogram
plt.hist(data['income'])
plt.title("Histogram")
plt.show()
```
